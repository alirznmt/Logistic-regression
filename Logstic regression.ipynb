```python
df=df[['Lag1','Lag2','Lag3','Lag4','Lag5','Volume','Direct']]
df
```

```python
x = df[['Lag1','Lag2','Lag3','Lag4','Lag5','Volume']]
y = df[['Direct']]
x =np.array(x)
y =np.array(y)
x , y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=0.2, )
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
x_train_with_bias = np.hstack((np.ones((x_train.shape[0],1)),x_train))
x_test_with_bias = np.hstack((np.ones((x_test.shape[0],1)),x_test))
```

```python
def sigmoid(c):
    return 1/(1 + np.exp(-c))
```

```python
def compute_gradient(X, y, weights):
    m = len(y)
    predictions = sigmoid(np.dot(X, weights))
    error = predictions - y
    gradient = np.dot(X.T, error) / m
    return gradient
```

```python
def mini_batch_sgd(X, y, learning_rate, batch_size, epochs):
    m, n = X.shape
    weights = np.random.randn(n, 1)
    step_size = learning_rate

    for epoch in range(epochs):
        indices = np.random.permutation(m)
        X_shuffled = X[indices]
        y_shuffled = y[indices]

        for i in range(0, m, batch_size):
            X_batch = X_shuffled[i:i+batch_size]
            y_batch = y_shuffled[i:i+batch_size]

            gradient = compute_gradient(X_batch, y_batch, weights)
            weights -= step_size * gradient

            # Decrease the step size using Robbins-Monro condition
            step_size = learning_rate / (epoch * m + i + 1)

    return weights
```

```python
learning_rate = 0.01
batch_size = 32
epochs = 200
```

```python
# Run mini-batch SGD
start_time = time.time()
weights_sgd = mini_batch_sgd(x_train_with_bias, y_train.reshape(-1, 1), learning_rate, batch_size, epochs)
duration = time.time() - start_time
print(f"Prediction duration: {duration:.6f} seconds")
```

```python
def logistic_regression(x, w):
    y_hat = sigmoid( x @ w )
    return y_hat
```

```python
#this function assign a class to instance
def check(array):
    for i in range(array.shape[0]):
        if array[i, 0] >= 0.5:
            array[i, 0] = 1
        elif array[i, 0] < 0.5 :
            array[i, 0] = 0
```

```python
def accuracy_logstic(y, y_hat):
    cnt = 0
    for i in range(len(y)):
        if y[i] == y_hat[i]:
            cnt += 1
    return cnt / len(y)
```

```python
yhat = logistic_regression(x_test_with_bias , weights_sgd)
check(yhat)
print(accuracy_logstic(y_test , yhat))
```

```python
def euclidean_distance( x1, x2):
        return np.sqrt(np.sum((x1 - x2) ** 2, axis=1))
```

```python


def knn_predict(X_train , y_train , X_test , k):
    y_pred = []
    for x_test in X_test:
        distances = euclidean_distance(X_train, x_test)
        k_indices = np.argsort(distances)[:k]
        k_labels = y_train[k_indices]
        unique_labels, counts = np.unique(k_labels, return_counts=True)
        y_pred.append(unique_labels[np.argmax(counts)])
    return np.array(y_pred)
```

```python
y_pred = knn_predict(x_train, y_train, x_test, k=3)

# Evaluate the accuracy
accuracy = np.mean(y_pred == y_test)
print(f"Accuracy: {accuracy}")
```

```python
momentum = 0.9
```

```python
def mini_batch_momentum(X, y, learning_rate, momentum, batch_size, epochs):
    m, n = X.shape
    weights = np.random.randn(n, 1)
    velocity = np.zeros((n, 1))

    for epoch in range(epochs):
        indices = np.random.permutation(m)
        X_shuffled = X[indices]
        y_shuffled = y[indices]

        for i in range(0, m, batch_size):
            X_batch = X_shuffled[i:i+batch_size]
            y_batch = y_shuffled[i:i+batch_size]

            gradient = compute_gradient(X_batch, y_batch, weights)
            velocity = momentum * velocity + learning_rate * gradient
            weights -= velocity

    return weights
```

```python
start_time1 = time.time()
weights_momentum = mini_batch_momentum(x_train_with_bias, y_train.reshape(-1, 1), learning_rate, momentum, batch_size, epochs)
duration1 = time.time() - start_time1
print(f"Prediction duration: {duration1:.6f} seconds")
yhat2 = logistic_regression(x_test_with_bias , weights_momentum)
check(yhat2)
print(accuracy_logstic(y_test , yhat2))
```
